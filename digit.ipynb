```python
#The MNIST database (Modified National Institute of Standards and Technology database) is a
#large database of handwritten digits that is commonly used for training various image processing systems.
#The database is also widely used for training and testing in the field of machine learning.
```

```python
#The data files train.csv and test.csv contain gray-scale images of hand-drawn digits, from zero through nine.
#Each image is 28 pixels in height and 28 pixels in width, for a total of 784 pixels in total.
#Each pixel has a single pixel-value associated with it, indicating the lightness or 
#darkness of that pixel, with higher numbers meaning darker. This pixel-value is an integer between 0 and 255, inclusive.
```

```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("C:/Users/user/Downloads/train.csv")
dataset
```

```python
cl=DecisionTreeClassifier()
```

```python
xtrain=dataset.iloc[0:21000,1:].values
print(xtrain)
xtrainl=dataset.iloc[0:21000,0].values
print(xtrainl)
```

```python
cl.fit(xtrain,xtrainl)
```

```python
xtest=dataset.iloc[21000:,1:].values
actuall=dataset.iloc[21000:,0].values
```

```python
d=xtest[16]
d.shape=(28,28)
plt.imshow(d,cmap="gray")
plt.show()
```

```python
print(cl.predict([xtest[16]]))
```

```python
#accuracy
p=cl.predict(xtest)
count=0
for i in range(0,21000):
    if (p[i]==actuall[i]):
        count+=1
print("accuracy is",(count/21000)*100)    
```

```python
cl.score(xtest,actuall)*100
```
